###Establish Dataframes

```python
import pandas as pd
df = pd.read_csv("Free Space - Sheet1 (2).csv")
cl = df[['LoanNumber', 'Error','Row']]
df.head()
```

###FeeType
Gets the fee type of the change(Feetype is listed in description variable of fees columns) from the corrective action

05 Rate Lock

```python
import re
Ft = []
for e in df['Error'].values:
  Ft.append(re.findall('"([^"]*)"', e)[0])
cl['ColumnName'] = Ft
cl['Row'] = df['Row']
cl['OldValue'] = df['RateLockDate']
value = []
for c in df['Corrective Action'].values:
  value.append(c)
cl['NewValue'] = value
cl.head()
```

```python
import re
Ft = []
for e in df['Error'].values:
  Ft.append(re.findall('"([^"]*)"', e)[1])
cl['FeeType'] = Ft
df['Fee Type'] = Ft
```

###ChangeAttribute and NewAtrributeValue
Gets the requested change variables from the corrective action column


```python
key = []
value = []
for e in df['Corrective Action'].values:
  #newvalue = e.split("update")[1].strip()
  #key.append(newvalue.split("=")[0])
  #value.append(newvalue.split("=")[1])
  key.append("SERVICEPROVIDERSOURCETYPE")
  value.append("BorrowerChosenProvider")
cl['ChangeAttribute'] = key
cl['NewAttributeValue'] = value
cl.head()
```

```python
cl.sort_values(by=[ 'LoanNumber', 'FeeType'], ascending=[True, False], inplace=True)
cl
```

### Return CSV

```python
from google.colab import files
cl.to_csv('output.csv', encoding = 'utf-8-sig', index=False)
files.download('output.csv')
```

```python
fees = []
for column in df:
  if column.startswith("FEE"):
    fees.append(column)
    df[column] = df[column].astype(str)
    df[column] = df[column].str.split('|')
df[fees].head()
```

```python
df.head()
```

```python
columns = []
for index, row in df.iterrows():
  for fee in fees:
    check = [x.upper() for x in row[fee]]
    if "DESCRIPTION={}".format(row['Fee Type'].upper()) in check or "TYPE={}".format(row['Fee Type'].upper()) in check:
      df.at[index,'Channel']  = fee
      df.at[index,'RESPAConformingYearType'] = row[fee][0]
      df.at[index,'Action Status'] = row[fee][-1]
      df.at[index,'FileType']  = '|'.join(row[fee])
      break

cl['Column'] = df['Channel']
cl['CellValue'] = df['FileType']
cl['IntegratedDisclosureSectionType'] = df['RESPAConformingYearType']
cl['ServiceProviderSourceType'] = df['Action Status']
cl.head()
```

```python
cl.to_csv('output.csv', encoding = 'utf-8-sig')
files.download('output.csv')
```
