```python
from google.colab import auth, drive
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

import pandas as pd

import os

gc = gspread.authorize(creds)
drive.mount('/content/drive')
folder = '/content/drive/MyDrive/TNT IDP/Manual Indexing'
```

#Grab Info from Template

```python
#opens selected workbook and worksheet for template
workbook = gc.open("Analytic DataPoints")
worksheet = workbook.worksheet("Sheet1")
#saves a dataframe of desired sheet
df = pd.DataFrame(worksheet.get_all_records(head=1))
```

#Add to worksheets

```python
#loops through each workbook in the folder
for file in os.listdir(folder):
  try:
    workbook = gc.open(file.replace(".gsheet", ""))
    #adds a sheet of selected name and size
    worksheet = workbook.worksheet("Manual Datapoints")
    workbook.del_worksheet( worksheet)
    workbook.add_worksheet(title="Manual Datapoints", rows="164", cols="7")
    worksheet = workbook.worksheet("Manual Datapoints")
    #adds calues to spreadsheet based on above created dataframe
    worksheet.update([df.columns.values.tolist()] + df.values.tolist())
  except:
    pass
```

#Update Column

```python
#opens selected workbook and worksheet for template
workbook = gc.open("Analytic DataPoints")
worksheet = workbook.worksheet("Sheet1")
#saves a dataframe of desired sheet
df = pd.DataFrame(worksheet.get_all_records(head=1))
array = df['Key to Look For'].values
```

```python
#loops through each workbook in the folder
for file in os.listdir(folder):
  try:
    workbook = gc.open(file.replace(".gsheet", ""))
    worksheet = workbook.worksheet("Manual Datapoints")
    cell_list = worksheet.range('D2:D164')

    for i, val in enumerate(array):  #gives us a tuple of an index and value
      cell_list[i].value = val    #use the index on cell_list and the val from cell_values
    worksheet.update_cells(cell_list)
  except:
    pass
```

#Update Cell

```python
for file in os.listdir(folder):
  try:
    sh = gc.open(file.replace(".gsheet", ""))
    worksheet = sh.worksheet("Manual Datapoints")
      #updates cell value
    worksheet.update('D1', 'Key to Look For')
  except:
    pass
```
